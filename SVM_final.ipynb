# Scaling

```python
X = df.iloc[:, : -1]
y = df.iloc[:, -1:]
```

```python
scaler = MinMaxScaler()

scaled_X = scaler.fit_transform(X)
scaled_y = scaler.fit_transform(y)

scaled_X.shape
```

# Train

```python
# split
train_X, test_X, train_y, test_y = train_test_split(scaled_X, scaled_y, test_size=0.2, random_state=0)
```

```python
# Create an SVM regressor
svm_regressor = SVR(kernel='rbf', gamma='scale', C=10, verbose=False)

# Train the model
svm_regressor.fit(train_X, train_y.ravel())
```
